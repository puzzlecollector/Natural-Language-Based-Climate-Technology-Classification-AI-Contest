```python
from catboost import CatBoostClassifier
from sklearn.model_selection import StratifiedKFold, KFold
import numpy as np 
import pandas as pd 
from tqdm import tqdm
```

```python
submission = pd.read_csv('./drive/MyDrive/dacon_NLP/sample_submission.csv') 
train = pd.read_csv('./drive/MyDrive/dacon_NLP/tfidf_train.csv') 
test = pd.read_csv('./drive/MyDrive/dacon_NLP/tfidf_test.csv') 

submission.shape, train.shape, test.shape
```

```python
train_labels = pd.read_csv("./drive/MyDrive/dacon_NLP/nlp_train.csv") 
train_labels = train_labels['label'] 

x_train = train 
y_train = train_labels 

x_train.shape, y_train.shape
```

```python
k = 10 
models = [] 
kfold = StratifiedKFold(n_splits = k, shuffle = True, random_state = 20210814) 
for n_fold, (train_idx, val_idx) in enumerate(kfold.split(x_train, y_train)): 
    print("Validating on fold {}".format(n_fold))

    train_x, val_x = x_train.loc[train_idx,:], x_train.loc[val_idx,:]
    train_y, val_y = y_train.loc[train_idx], y_train.loc[val_idx] 
    
    model = CatBoostClassifier(iterations = 50000, 
                               learning_rate = 0.01, 
                               loss_function = 'MultiClass',
                               eval_metric = 'Accuracy',
                               use_best_model = True, 
                               task_type = 'GPU',
                               verbose = 100)  
    
    
    model.fit(train_x, train_y, eval_set = (val_x, val_y))  
    # save model just in case 
    model.save_model('./drive/MyDrive/dacon_NLP/catboost_' + str(n_fold)) 
    models.append(model)
```

```python
pred1 = models[0].predict_proba(test) 
pred2 = models[1].predict_proba(test) 
pred3 = models[2].predict_proba(test) 
pred4 = models[3].predict_proba(test) 
pred5 = models[4].predict_proba(test)
pred6 = models[5].predict_proba(test) 
pred7 = models[6].predict_proba(test) 
pred8 = models[7].predict_proba(test) 
pred9 = models[8].predict_proba(test) 
pred10 = models[9].predict_proba(test)


pred_avg = (pred1 + pred2 + pred3 + pred4 + pred5 + pred6 + pred7 + pred8 + pred9 + pred10)/10.0


np.save('./drive/MyDrive/dacon_NLP/catboost_10_fold.npy', pred_avg)
```
